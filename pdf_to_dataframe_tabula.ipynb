```python
import tabula
import pandas as pd
```

```python
df = tabula.read_pdf("TransactionsList.pdf",stream=False,guess=True,lattice=True)
df
```

```python
df[0]
```

```python
df_1 = df[0].copy()
df_1
```

```python
df_1 = df_1.iloc[:5]
df_1
```

```python
df_1b = df_1[df_1.columns[2:4]].dropna(how="all",axis=0).transpose()
df_1b.columns = df_1b.iloc[0]
df_1b = df_1b[1:]
df_1b.reset_index(inplace=True)
df_1b.rename(columns={"index":"Numero_transazione"},inplace=True)
df_1b
```

```python
df_1 = df_1[df_1.columns[:2]].transpose()
df_1.columns = df_1.iloc[0]
df_1 = df_1[1:]
df_1.reset_index(inplace=True)
df_1.rename(columns={"index":"Data_orario","Data contabile":"Data"},inplace=True)
df_1
```

```python
df_1 = pd.concat([df_1,df_1b],axis=1)
df_1
```

```python
df_2 = df[1].copy()
df_2
```

```python
df_2 = df_2.iloc[:-1].copy()
df_2.dropna(how="all",axis=1, inplace=True)
df_2.rename(columns={
    "Articolo":"Categoria_articolo",
    "Quantità":"Nome_articolo",
    "Importo":"Quantity",
    "IVA":"Importo_lordo",
    "Unnamed: 0":"IVA"
    },
    inplace=True)
df_2
```

```python
df_transaction = pd.concat([df_1,df_2],axis=1)
df_transaction.fillna(method="ffill",inplace=True)
df_transaction['Categoria_articolo'] = df_transaction["Categoria_articolo"].apply(lambda x : x.replace("\r",""))
df_transaction['Nome_articolo'] = df_transaction["Nome_articolo"].apply(lambda x : x.replace("\r",""))
df_transaction['Data_orario'] = df_transaction["Data_orario"].apply(lambda x :pd.to_datetime(x))
df_transaction['Data'] = df_transaction["Data"].apply(lambda x :pd.to_datetime(x))
df_transaction['Ospiti'] = df_transaction["Ospiti"].apply(lambda x : int(x))
df_transaction['Carta'] = df_transaction["Carta"].fillna(value="NO CARTA")
df_transaction
```

```python
df_transaction.to_csv("final_result.csv")
```
